```python
arr = [1096770097, 1952395366, 1600270708, 1601398833, 1716808014, 1734293090,  943076964, 946090341]
res = []
flag = ""
for i in arr:
    x = format(i,'032b')   
    for j in range(0,4):
        res.append(x[8*j:(8*j+8)])

        
for byte in res:
    n = int(byte,2)
    if n in range(32,127):
        flag = flag + chr(n)
print(f"picoCTF{{{flag}}}")
```
