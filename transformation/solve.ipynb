Nhận xét 1: [chr((ord(flag[i]) << 8) + ord(flag[i + 1])) for i in range(0, len(flag), 2)] = enc

Nhận xét 2: với mỗi ký tự thuộc enc ta tiến hành phân tích ra chr(i) và chr(i+1) và nhận thấy cặp [chr(i);chr(i+1)] là duy nhất đối vơi mỗi enc[i_tương ứng]

```python
res = ""
enc = "灩捯䍔䙻ㄶ形楴獟楮獴㌴摟潦弸形㝦㘲捡㕽"
for target in enc:
    for i in range(32,127):
        for j in range(32,127):
            if target == chr(( i << 8) + j):
               
               res_next = chr(i)
               res_next = res_next + chr(j)
               res = res + res_next
print(res)
```

Cách khác : enc[i] = chr((ord(flag[i]) << 8) + ord(flag[i + 1])) suy ra enc[i] là số nguyên 16-bit với byte cao biểu diễn flag[i] và byte thấp biểu diễn flag[i+1] (vì flag[i] được dịch trái, nghĩa là được thêm 8 số 0 vào cuối).

Vì vậy, flag[i] = chr(ord(enc[i]) >> 8) (cắt 8  số 0  ở cuối) và flag[i+1] = chr(ord(enc[i])& 0xFF) , 0xFF = 0000 0000 1111 1111
