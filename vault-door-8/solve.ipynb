- Đây là bài làm rối (Obfuscation) ở mức giới thiệu, vì nó làm rối với mức ý nghĩa trong bảo vệ thấp ở đây ta dùng từ scramble thay vì obfuscation, vì : 

 * Làm rồi bằng cách làm xấu format code --> ta có thể tự động format code bằng IDE;

 * Làm rối logic code bằng cách sử dụng comment nhiễu (/* c = ....*/) --> ở đây ngay commen đầu tiên c = switch Bits(c,14,3) và vì type(c) là chả ~ 1 bytes (8 bit) nên không có bit thứ 13 và ở ngay comment dưới có thêm d = ... và e= ... mà chẳng liên quan gì đến hàm scramble(...) --> từ đây tôi mạnh dạn suy luận là comment ở đây chỉ gây nhiễu dù đây không phải suy luận chặt;

 * Chính hàm switchBits làm rối vì nếu ta lao vào phân tích và có đảo ngược logic hàm sẽ sai và việc ta cần làm duy nhất là đảo logic của hàm scramble và giữ nguyên logic của hàm switchBits.

```python
def switchBits(c,p1,p2) :
        x = ord(c)
        mask1 = (1 << p1)
        mask2 = (1 << p2)
        bit1 = (x & mask1)
        bit2 = (x & mask2)
        rest =  (x & ~(mask1 | mask2))
        shift =  (p2 - p1)
        result =((bit1 << shift) | (bit2 >> shift) | rest)
        return chr(result)

def DeScramble(password) :
        a =list(password)
        for b in range(0,len(a)):
            c = a[b]
            c = switchBits(c, 6, 7)
            c = switchBits(c, 2, 5)
            c = switchBits(c, 3, 4)
            c = switchBits(c, 0, 1)
            c = switchBits(c, 4, 7)
            c = switchBits(c, 5, 6)
            c = switchBits(c, 0, 3)
            c = switchBits(c, 1, 2)                                     
            
            a[b] = c
        
        return a
password = [0xF4, 0xC0, 0x97, 0xF0, 0x77, 0x97, 0xC0, 0xE4, 0xF0, 0x77, 0xA4, 0xD0, 0xC5, 0x77, 0xF4, 0x86, 0xD0, 0xA5, 0x45, 0x96, 0x27, 0xB5, 0x77, 0x94, 0xC1, 0x95, 0xE0, 0xA4, 0xA5, 0x95, 0xF0, 0xE0]
key = [chr(i) for i in password]

flag = ""
key = DeScramble(key)
for i in key:
    flag = flag + i
print(f"picoCTF{{{flag}}}")

```
