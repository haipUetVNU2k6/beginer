Thử thách WinAntiDbg0x100 chủ yếu giới thiệu hàm IsDebugPresent, đây là hàm anti-debug.

Tóm tắt, hàm main kiểm tra bạn/ ép bạn dùng công cụ debug nhưng khối sau đó thì lại anti-debug.Nhiệm vụ của bạn là không đi vào khối chứ IsDebugPresent hoặc thay đổi cờ để đánh lừa chương trình.Ở đây chúng ta dùng cách thứ 2 bằng cách thay đổi cờ Zero Flag(ZF) từ 0x0 sang 0x1. 

Tóm lại, ở thử thách giúp bạn làm quen công cụ IDA và Detect-It-Easy, cụ thể là quen với graph flow: biết mỗi bảng/khối chứa gì? nhiệm vụ là gì? các đường nối mũi tên có ý nghĩa ra sao? tổng quan chương trình như thế nào? Sinh mã psedoucode(ngôn ngữ C) như thế nào?
